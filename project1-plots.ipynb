```python
import numpy as np
import matplotlib.pyplot as plt
nodes = 1
cores = 1
paths = [f"init_{nodes}_{cores}.txt", f"mid_{nodes}_{cores}.txt", f"end_{nodes}_{cores}.txt"]
```

```python
file_path = paths[0]
data = np.loadtxt(file_path)
max = data.max()

plt.imshow(data.T, cmap='plasma', origin='upper', vmin = 0, vmax = 1, extent = [-0.5, 0.5, -0.5, 0.5])
cbar = plt.colorbar()
cbar.set_label('C(x,y)', rotation=270, labelpad=15)
plt.xlabel('x')
plt.ylabel('y')
plt.title("t = 0")
plt.show()
```
